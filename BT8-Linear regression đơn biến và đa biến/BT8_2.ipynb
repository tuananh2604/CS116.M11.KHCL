```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("50_Startups.csv")
data
```

```python
transformer = ColumnTransformer([('encoder', OneHotEncoder(), [3])], remainder='passthrough')
```

```python
data = np.array(transformer.fit_transform(data), dtype = np.str)
```

```python
df = pd.DataFrame(data,columns=['Califonia_State','Florida_State','NewYork_State','R&D Spend','Administration','Marketing Spend','Profit'])
```

```python
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
reg = LinearRegression().fit(X_train[['Califonia_State','Florida_State','NewYork_State','R&D Spend','Administration','Marketing Spend']],y_train)
reg.score(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```
