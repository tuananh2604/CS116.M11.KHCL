```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('hinh.jpg',0)
cv2_imshow(image)
```

```python
image.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(100)
image2 = pca.fit(image)
components = pca.transform(image)
```

```python
components.shape
```

```python
recovered_image = pca.inverse_transform(components)
cv2_imshow(recovered_image)
```
